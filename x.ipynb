```python
import os
import pandas as pd
from datetime import datetime
import re
def connect_angle_to_frame(folder_path, MR_file, nexus_file_array, nexus_index_file):
    # Bước 1: Đọc dữ liệu từ CSV1 để lấy frame, bắt đầu từ hàng 6
    df_MR = pd.read_csv(os.path.join(folder_path, MR_file))   
    nexus_file_with_number = nexus_file_array[nexus_index_file]
    # df_Nexus = pd.read_csv(os.path.join(folder_path, nexus_file_with_number), skiprows=5)  
    def read_csv_with_skiprows(path):
        filename = os.path.basename(path)
        if re.search(r'\d', filename):
            skiprows = 5
        else:
            skiprows = 4
        df = pd.read_csv(path, skiprows=skiprows)
        return df

    df_Nexus= read_csv_with_skiprows(os.path.join(folder_path, nexus_file_with_number))

    # Giả sử thời gian bắt đầu của CSV1 ở hàng đầu tiên và của CSV3 ở hàng thứ 3 (dựa trên thông tin trước đó)
    MR_time_start = df_MR.iloc[:, 11].iloc[0]  # Lấy thời gian từ hàng 1
    Nexus_time_start = df_MR.iloc[:, 11].iloc[nexus_index_file + 2]  # Lấy thời gian từ hàng 3

    # Chuyển đổi thời gian thành datetime
    fmt = "%Y-%m-%dT%H:%M:%S.%fZ"
    MR_time = datetime.strptime(MR_time_start, fmt)
    Nexus_time = datetime.strptime(Nexus_time_start, fmt)

    # Tính độ lệch thời gian giữa CSV1 và CSV3
    time_difference = Nexus_time - MR_time
    time_difference_seconds = time_difference.total_seconds()

    # Tính frame bắt đầu trong CSV3
    start_frame_on_nexus = int(time_difference_seconds / 0.01)
    start_frame_on_nexus_to_MR = start_frame_on_nexus + 5

    start_from_cutted_frame = df_Nexus.iloc[0, 0]
    start_frame_on_nexus_to_MR = start_frame_on_nexus_to_MR + int(start_from_cutted_frame)

    # Bước 2: Tìm chỉ số của hàng chứa "Trajectories"
    trajectories_index = df_Nexus.index[df_Nexus.iloc[:, 0] == 'Trajectories'].tolist()

    if trajectories_index:
        # Nếu tìm thấy, chỉ giữ lại các dòng trước chỉ số của hàng "Trajectories"
        df_Nexus = df_Nexus.iloc[:trajectories_index[0]]  # Giữ lại tất cả các dòng trước hàng "Trajectories"

    frames_MR  = df_MR.iloc[start_frame_on_nexus_to_MR:, 0].reset_index(drop=True)
    columns_from_nexus = df_Nexus.iloc[0:, [0, 11]].reset_index(drop=True)

    # Lấy chiều dài của df_Nexus để cắt DataFrame mới
    length_of_nexus = len(columns_from_nexus)

    # Cắt DataFrame mới theo chiều dài của nexus file
    conect_frame_angle = pd.concat([frames_MR[:length_of_nexus], columns_from_nexus[:length_of_nexus]], axis=1)

    # Đặt tên cho các cột nếu cần
    conect_frame_angle.columns = ['frame', 'index_on_nexus', 'angle']
    # Loại bỏ các hàng có giá trị NaN trong cột frame trước khi chuyển đổi
    conect_frame_angle = conect_frame_angle.dropna(subset=['frame'])
    conect_frame_angle['frame'] = ((conect_frame_angle['frame'].astype(float)) * 100).astype(int)

    return conect_frame_angle

def create_combined_dataframe(folder_path, MR_file, nexus_file_array):
    combined_dataframe = pd.DataFrame()  # Khởi tạo một DataFrame rỗng

    # Lặp qua từng chỉ số trong nexus_file_array
    for nexus_index_file in range(len(nexus_file_array)):
        # Gọi hàm connect_angle_to_frame và tạo DataFrame
        df = connect_angle_to_frame(folder_path, MR_file, nexus_file_array, nexus_index_file)
        
        # Nối các DataFrame theo chiều dọc
        combined_dataframe = pd.concat([combined_dataframe, df])
    
    return combined_dataframe

# Gọi hàm để tạo DataFrame kết hợp
folder_path = "C:\\Users\\gochu\\Downloads\\Data Khuong running\\"
MR_file = 'MR32024-10-31-10-36_Treadmill_5.0_kph.csv'
nexus_file_array = [
    'Khuong runnning.csv',
    'Khuong runnning01.csv',
    'Khuong runnning02.csv',
    'Khuong runnning03.csv',
    'Khuong runnning04.csv',
    'Khuong runnning05.csv',
    'Khuong runnning06.csv',
]

combined_dataframe = create_combined_dataframe(folder_path, MR_file, nexus_file_array)

# In kết quả
print(combined_dataframe)
```
